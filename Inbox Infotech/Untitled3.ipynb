```python
import pandas as pd
import numpy as np
```

```python
file_path = "C:/Users/Tisha/Downloads/Car(final)Table.xlsx"
car_df = pd.read_excel(file_path)
car_df.info()
```

```python
def assign_rating(row):
    rating = 2.5  # Default base rating

    # Higher ratings for newer cars
    if row['Year of Manufacture'] > 2012:
        rating += 1.0  # Increase rating
    elif row['Year of Manufacture'] <= 2012:
        rating += 0.5

    # Car Type influence
    luxury_cars = ['Luxury', 'Sports', 'Convertible']
    suv_sedan = ['SUV', 'Sedan']
    
    if row['Car Type'] in luxury_cars:
        rating += 1.2
    elif row['Car Type'] in suv_sedan:
        rating += 0.5
    else:
        rating +=0.3
    # Mileage impact
    if row['Mileage (km/l)'] > 18:
        rating += 0.3
    elif row['Mileage (km/l)'] < 12:
        rating -= 0.2

    # Price-based adjustment
    if row['Price per Hour (INR)'] > 200 and row['Car Type'] not in luxury_cars:
        rating -= 0.6  # Expensive non-luxury cars have slightly lower ratings

    # Ensure rating is within 1.0 to 5.0
    return round(np.clip(rating, 1.0, 5.0), 1)

# Apply rating function to each row
car_df['Rating'] = car_df.apply(assign_rating, axis=1)
```
