```python
import numpy as np

# Create a random number generator with a fixed seed for reproducibility
rng = np.random.default_rng(42)

# Dictionary to track assigned cars per (Pickup_Location, rental_date)
assigned_cars = {}

def get_unique_car_id(row):
    key = (row['Pickup_Location'], row['rental_date'])
    
    # Get available cars for the given location
    available_cars = combined_data[combined_data['City'] == row['Pickup_Location']]['Car_Id'].values
    
    # Remove already assigned cars for the same (location, date)
    available_cars = [c for c in available_cars if c not in assigned_cars.get(key, set())]
    
    # If all cars are assigned, allow reuse by resetting the assigned list
    if not available_cars:
        assigned_cars[key] = set()  # Reset assigned cars (allow reuse)
        available_cars = combined_data[combined_data['City'] == row['Pickup_Location']]['Car_Id'].values

    # Assign a car if available
    if available_cars:
        selected_car = rng.choice(available_cars)
        assigned_cars.setdefault(key, set()).add(selected_car)
        return selected_car
    return None  # No available cars

# Assign car_id while allowing reuse across different dates
hotel['Car_ID'] = hotel.apply(get_unique_car_id, axis=1)

# Check for duplicates
duplicates = hotel.duplicated(subset=['Pickup_Location', 'rental_date', 'Car_ID']).sum()
print(f"Number of duplicate (Pickup_Location, rental_date, car_id) entries after fix: {duplicates}")
```

```python
def calculate_total_amount(row):
    car_info = combined_data[combined_data['Car_Id'] == row['Car_ID']]# match the car_table id with the car_rental_id
    if not car_info.empty:
        return (car_info['Base_Fare'].values[0] + car_info['Price per day'].values[0]) * row['Duration']
    return None

hotel['total_amount'] = hotel.apply(calculate_total_amount, axis=1)
```
