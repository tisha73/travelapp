```python
import random

arr_drop_loc = pd.unique(ren_car['dropoffLocation'])
arr_pick_loc = pd.unique(ren_car['pickupLocation'])
arr_car_type = pd.unique(ren_car['carType'])
arr_car_rental_agency = pd.unique(ren_car['rentalAgency'])
arr_car_rental_duration = pd.unique(ren_car['rentalDuration'])
arr_fuelPolicy_car = pd.unique(ren_car['fuelPolicy'])
arr_bookingStatus_car = pd.unique(ren_car['bookingStatus'])
```

```python
import numpy as np

np.random.seed(42)  

User_merged_df['dropoffLocation_car'] = np.random.choice(arr_drop_loc, size=len(User_merged_df))
User_merged_df['pickupLocation_car'] = np.random.choice(arr_pick_loc, size=len(User_merged_df))
User_merged_df['car_Type'] = np.random.choice(arr_car_type, size=len(User_merged_df))
User_merged_df['Car_rentalAgency'] = np.random.choice(arr_car_rental_agency, size=len(User_merged_df))
User_merged_df['Car_rentalDuration'] = np.random.choice(arr_car_rental_duration, size=len(User_merged_df))
User_merged_df['Car_fuelPolicy'] = np.random.choice(arr_fuelPolicy_car, size=len(User_merged_df))
User_merged_df['Car_bookingStatus'] = np.random.choice(arr_bookingStatus_car, size=len(User_merged_df))
```
